```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import SpectralClustering
cluster=SpectralClustering(n_clusters=2,assign_labels='discretize',random_state=0)
cluster.fit_predict(x)
labels=cluster.labels_
```

```python
supervised=dataset
supervised['Cluster_group']=labels
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```

```python
from sklearn.cluster import SpectralClustering
cluster=SpectralClustering(affinity='rbf', assign_labels='kmeans', coef0=1, degree=3,
                   eigen_solver=None, eigen_tol=0.0, gamma=1.0,
                   kernel_params=None, n_clusters=8, n_init=10, n_jobs=None,
                   n_neighbors=10, random_state=None)
cluster.fit_predict(x)
labels=cluster.labels_
```

```python
supervised=dataset
supervised['Cluster_group']=labels
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
